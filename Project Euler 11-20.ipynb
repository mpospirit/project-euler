### Highly divisible triangular number

#### Problem 12
The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 1: 1
 3: 1,3
 6: 1,2,3,6
10: 1,2,5,10
15: 1,3,5,15
21: 1,3,7,21
28: 1,2,4,7,14,28
We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
np.warnings.filterwarnings('ignore', category=np.VisibleDeprecationWarning)
```

```python
def prime_factors(n):
    factors = []
    if n > 1:  # this is for stopping recursive process
        i = 2
        while n % i != 0:  # increasing the counter until reaching an exact divisor
            i += 1
        factors.append(i)  # appending divisor to the list
        factors.extend(prime_factors(n / i))  # repeating the process
    return factors


def unique_divisors(number):
    lis = prime_factors(number)
    my_array = []

    for i in range(1, len(lis) + 1):
        my_array.append(list(combinations(lis, i)))

    flat_list = [item for sublist in my_array for item in sublist]
    flat_list = np.array(flat_list)

    df = pd.DataFrame(flat_list)

    prod = []

    for i in df[0]:
        prod.append(np.prod(i))

    df["Prod"] = prod
    return df


def triangular(k):
    i = 0
    num = 0
    while i < k + 1:
        num += i
        i += 1
    return num


iterator = 0
# graph=[]

while True:
    result = unique_divisors(triangular(iterator))
    print(iterator, triangular(iterator), result)
    #     graph.append(result)
    if result >= 500:
        print(triangular(iterator))
        break
    iterator += 1

# plt.figure(figsize=(16,10))
# plt.plot(graph)
```

### Longest Collatz sequence

#### Problem 14
The following iterative sequence is defined for the set of positive integers:

n → n/2 (n is even)
n → 3n + 1 (n is odd)

Using the rule above and starting with 13, we generate the following sequence:

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
It can be seen that this sequence (starting at 13 and finishing at 1) contains 10 terms. Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at 1.

Which starting number, under one million, produces the longest chain?

NOTE: Once the chain starts the terms are allowed to go above one million.

```python
def collatz(number):
    seq=[]
    while number!=1:
        seq.append(number)
        if number%2==0:
            number=number/2
        else:
            number=3*number+1
    seq.append(1)
    
    return {'Sequence': seq,'Length': len(seq)}

longest=2

for i in range(2,1000000):
    if collatz(i)['Length']>collatz(longest)['Length']:
        longest=i

collatz(longest)
```

### Lattice paths
 
#### Problem 15
Starting in the top left corner of a 2×2 grid, and only being able to move to the right and down, there are exactly 6 routes to the bottom right corner.

![](https://projecteuler.net/project/images/p015.png)

How many such routes are there through a 20×20 grid?

```python
# I need to go left 20 times and go down 20 times
# It is a simple permutation problem that we need to arrange 20 Ls and 20 Ds

np.math.factorial(40) / (np.math.factorial(20) * np.math.factorial(20))
```

75

```python
triangle = [[75],
            [95,64],
            [17,47,82],
            [18,35,87,10],
            [20, 4,82,47,65],
            [19, 1,23,75, 3,34],
            [88, 2,77,73, 7,63,67],
            [99,65, 4,28, 6,16,70,92],
            [41,41,26,56,83,40,80,70,33],
            [41,48,72,33,47,32,37,16,94,29],
            [53,71,44,65,25,43,91,52,97,51,14],
            [70,11,33,28,77,73,17,78,39,68,17,57],
            [91,71,52,38,17,14,91,43,58,50,27,29,48],
            [63,66, 4,68,89,53,67,30,73,16,69,87,40,31],
            [ 4,62,98,27,23, 9,70,98,73,93,38,53,60, 4,23]]


def max_path(triangle):
    for i in range(len(triangle)-2, -1, -1):
        for j in range(len(triangle[i])):
            triangle[i][j] += max(triangle[i+1][j], triangle[i+1][j+1])
    return triangle[0][0]

max_path(triangle)
```

### Counting Sundays

#### Problem 19

How many Sundays fell on the first of the month during the twentieth century (1 Jan 1901 to 31 Dec 2000)?

```python
sundays = 0
day = 1

for year in range(1901, 2001):
    for month in range(1, 13):
        if datetime.datetime(year, month, day).strftime("%A") == "Sunday":
            sundays += 1

sundays
```
